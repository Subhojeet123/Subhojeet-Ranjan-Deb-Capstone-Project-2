## 2.3 Removing Unnecessary Columns

### Task

Remove features containing an excessive number of missing values to improve data quality and reduce unnecessary complexity.

---

### Explanation

Attributes with a very high percentage of missing values often contribute little useful information for predictive modelling.

Removing these features reduces dataset dimensionality, simplifies preprocessing, and improves computational efficiency while preserving the most informative variables.

```python
# Calculate percentage of missing values
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Columns with more than 50% missing values
columns_to_drop = missing_percentage[missing_percentage > 50].index

print("Number of columns to remove:", len(columns_to_drop))

# Remove columns
df = df.drop(columns=columns_to_drop)

print("Dataset Shape after removing columns:", df.shape)

# Fill remaining missing values
df = df.fillna(df.median(numeric_only=True))

print("\nRemaining Missing Values:")
print(df.isnull().sum().sum())
```

### Interpretation

The selected features containing excessive missing values were successfully removed from the dataset.

This reduced the dimensionality of the dataset while retaining the most relevant sensor measurements required for further analysis.

# Question 4: Data Pre-processing

## Task

Prepare the cleaned dataset for machine learning by selecting the input features and target variable, encoding the target labels, splitting the dataset into training and testing sets, standardizing the numerical features, and handling class imbalance using the Synthetic Minority Over-sampling Technique (SMOTE).

---

## Explanation

Data preprocessing is one of the most important stages in a machine learning workflow. Raw data must be transformed into a suitable format before training predictive models.

In this section, the input features and target variable are separated, the target labels are encoded into numerical values, the dataset is divided into training and testing subsets, numerical features are standardized, and class imbalance is addressed using SMOTE. These preprocessing steps improve model stability, learning efficiency, and predictive performance.

## 4.1 Feature Selection

### Task

Separate the predictor variables (sensor measurements) from the target variable (Pass/Fail).

---

### Explanation

The predictor variables contain the numerical sensor measurements collected during the semiconductor manufacturing process, while the target variable indicates whether a product successfully passed or failed quality inspection.

Separating the features and target variable prepares the dataset for machine learning model development.

```python
# Features
X = df.drop(columns=["Time", "Pass/Fail"])

# Target
y = df["Pass/Fail"]

print("Feature Matrix Shape :", X.shape)
print("Target Vector Shape  :", y.shape)

print("\nTarget Classes:")
print(y.value_counts())
```

### Interpretation

The dataset was successfully divided into predictor variables and the target variable.

The feature matrix contains all sensor measurements used for model training, while the target vector represents the classification labels indicating product quality.

## 4.2 Target Variable Encoding

### Task

Convert the target variable into numerical labels suitable for machine learning algorithms.

---

### Explanation

Most machine learning algorithms require categorical labels to be represented as numerical values.

The original target labels are encoded into binary values, where the majority and minority classes can be processed efficiently during model training.

```python
# Encode target variable
y = y.replace({
    -1: 0,
     1: 1
})

print("Encoded Target Distribution:\n")
print(y.value_counts())
```

### Interpretation

The target variable was successfully encoded into binary numerical values.

This transformation enables machine learning algorithms to interpret the class labels correctly while preserving the original classification information.

## 4.3 Train-Test Split

### Task

Divide the dataset into training and testing subsets for model development and evaluation.

---

### Explanation

The dataset is divided into separate training and testing sets to evaluate how well the machine learning models generalize to unseen data.

The training set is used to learn patterns from the data, while the testing set provides an unbiased evaluation of model performance.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print("\nTraining Labels :", y_train.shape)
print("Testing Labels  :", y_test.shape)
```

### Interpretation

The dataset was successfully divided into training and testing subsets.

Using separate datasets for training and evaluation helps prevent overfitting and provides a more reliable assessment of the machine learning models.

## 4.4 Feature Standardization

### Task

Standardize the numerical sensor features so that they have a mean of zero and a standard deviation of one.

---

### Explanation

The semiconductor manufacturing dataset contains sensor measurements with different numerical scales and units. Features with larger numerical values may disproportionately influence certain machine learning algorithms, particularly distance-based models such as Support Vector Machines (SVM) and Logistic Regression.

Feature standardization transforms each numerical feature to have a mean of zero and a standard deviation of one. This ensures that all sensor measurements contribute equally during model training, leading to improved convergence, stability, and predictive performance.

```python
from sklearn.preprocessing import StandardScaler

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the training and testing datasets
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Training Data Shape :", X_train_scaled.shape)
print("Testing Data Shape  :", X_test_scaled.shape)

print("\nFeature Standardization Completed Successfully.")
```

### Interpretation

The numerical sensor features were successfully standardized using the StandardScaler technique.

After standardization, each feature has a comparable scale, preventing variables with larger numerical ranges from dominating the learning process. This preprocessing step is particularly important for machine learning algorithms that rely on distance calculations and gradient-based optimization.

## 4.5 Handling Class Imbalance Using SMOTE

### Task

Balance the training dataset using the Synthetic Minority Over-sampling Technique (SMOTE) to improve the learning capability of machine learning models.

---

### Explanation

The target variable in this dataset is highly imbalanced, with significantly more "Pass" observations than "Fail" observations. Training a model on such an imbalanced dataset may result in poor detection of the minority class.

SMOTE (Synthetic Minority Over-sampling Technique) addresses this problem by generating synthetic samples for the minority class rather than simply duplicating existing observations. This creates a more balanced training dataset and helps machine learning models better recognize defective semiconductor products.

```python
# Install imbalanced-learn if required
# Uncomment the next line only if SMOTE is not installed
# !pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE only to the training data
X_train_smote, y_train_smote = smote.fit_resample(
    X_train_scaled,
    y_train
)

print("Before SMOTE")
print(y_train.value_counts())

print("\nAfter SMOTE")
print(y_train_smote.value_counts())

print("\nTraining Feature Shape After SMOTE:", X_train_smote.shape)
print("Training Target Shape After SMOTE :", y_train_smote.shape)
```

### Interpretation

SMOTE successfully balanced the training dataset by generating synthetic samples for the minority class.

After oversampling, both classes contain an equal number of observations, reducing class imbalance and improving the ability of machine learning models to learn patterns associated with defective semiconductor products.

This preprocessing step is expected to improve classification performance, particularly for the minority class, during model evaluation.

Train multiple supervised machine learning classification models using the preprocessed semiconductor manufacturing dataset and evaluate their predictive performance.

## 5.1 Logistic Regression

### Task

Train a Logistic Regression classifier using the balanced training dataset and evaluate its performance on the testing dataset.

---

### Explanation

Logistic Regression is a widely used supervised learning algorithm for binary classification problems. It estimates the probability that an observation belongs to a particular class using the logistic (sigmoid) function.

Despite its simplicity, Logistic Regression often serves as a strong baseline model for evaluating classification performance.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report
)

# Train Logistic Regression
lr_model = LogisticRegression(random_state=42)

lr_model.fit(X_train_smote, y_train_smote)

# Predictions
y_pred_lr = lr_model.predict(X_test_scaled)

# Accuracy
lr_accuracy = accuracy_score(y_test, y_pred_lr)

print("Logistic Regression Accuracy:")
print(round(lr_accuracy * 100, 2), "%")

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_lr))

print("\nClassification Report")
print(classification_report(y_test, y_pred_lr))
```

### Interpretation

The Logistic Regression model achieved an overall accuracy of **87.50%** on the testing dataset.

The confusion matrix indicates that the model correctly classified most products belonging to the majority class. However, its ability to detect defective semiconductor products (minority class) remains limited, as reflected by the relatively low recall and F1-score.

Although Logistic Regression provides a reliable baseline, additional models are evaluated to improve minority class prediction.

## 5.2 Random Forest Classifier

### Task

Train a Random Forest classifier and evaluate its classification performance using the testing dataset.

---

### Explanation

Random Forest is an ensemble learning algorithm that combines multiple decision trees to improve predictive accuracy and reduce overfitting.

Each decision tree is trained on a random subset of the data, and the final prediction is obtained through majority voting. This approach generally produces robust and accurate classification models.

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_model.fit(X_train_smote, y_train_smote)

# Make predictions
y_pred_rf = rf_model.predict(X_test_scaled)

# Accuracy
rf_accuracy = accuracy_score(y_test, y_pred_rf)

print("Random Forest Accuracy:")
print(f"{rf_accuracy * 100:.2f}%")

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_rf))

print("\nClassification Report")
print(classification_report(y_test, y_pred_rf))
```

### Interpretation

The Random Forest classifier achieved an overall accuracy of **93.38%**, which is the highest among the evaluated models.

The model classified nearly all normal semiconductor products correctly. However, despite its excellent overall accuracy, it identified only a small number of defective products.

This demonstrates that high accuracy alone may not indicate the best model when dealing with imbalanced datasets. Metrics such as recall and F1-score for the minority class should also be considered.

## 5.3 Support Vector Machine (SVM)

### Task

Train a Support Vector Machine classifier and evaluate its predictive performance.

---

### Explanation

Support Vector Machine (SVM) is a powerful supervised learning algorithm that identifies an optimal decision boundary to maximize the separation between different classes.

SVM performs particularly well on high-dimensional datasets and often provides strong classification performance after feature standardization.

```python
from sklearn.svm import SVC

# Initialize the SVM model
svm_model = SVC(
    kernel="rbf",
    random_state=42
)

# Train the model
svm_model.fit(X_train_smote, y_train_smote)

# Make predictions
y_pred_svm = svm_model.predict(X_test_scaled)

# Accuracy
svm_accuracy = accuracy_score(y_test, y_pred_svm)

print("Support Vector Machine Accuracy:")
print(f"{svm_accuracy * 100:.2f}%")

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_svm))

print("\nClassification Report")
print(classification_report(y_test, y_pred_svm))
```

### Interpretation

The Support Vector Machine classifier achieved an overall accuracy of **92.28%**.

The model performed well in identifying products belonging to the majority class but failed to correctly classify defective semiconductor products in the testing dataset.

This result indicates that, despite strong overall accuracy, SVM is less effective for minority class prediction in this highly imbalanced classification problem.

## 5.4 Gaussian Naive Bayes

### Task

Train a Gaussian Naive Bayes classifier and evaluate its performance on the testing dataset.

---

### Explanation

Gaussian Naive Bayes is a probabilistic classification algorithm based on Bayes' Theorem. It assumes that the predictor variables are conditionally independent and normally distributed within each class.

Although computationally efficient, its simplifying assumptions may reduce predictive performance when feature independence does not hold.

```python
from sklearn.naive_bayes import GaussianNB

# Initialize the model
gnb_model = GaussianNB()

# Train the model
gnb_model.fit(X_train_smote, y_train_smote)

# Predictions
y_pred_gnb = gnb_model.predict(X_test_scaled)

# Accuracy
gnb_accuracy = accuracy_score(y_test, y_pred_gnb)

print("Gaussian Naive Bayes Accuracy:")
print(f"{gnb_accuracy * 100:.2f}%")

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_gnb))

print("\nClassification Report")
print(classification_report(y_test, y_pred_gnb))
```

### Interpretation

The Gaussian Naive Bayes classifier achieved an overall accuracy of **33.82%**, the lowest among all evaluated models.

Although the model identified a relatively large proportion of defective products, it misclassified many normal products, resulting in poor overall performance.

This indicates that the independence assumption of Gaussian Naive Bayes is not well suited to the complex relationships among semiconductor sensor measurements.

# Question 6: Cross Validation

## Task

Evaluate the stability and generalization performance of the selected machine learning model using K-Fold Cross Validation.

---

## Explanation

Cross Validation is a robust model evaluation technique that assesses how well a machine learning model generalizes to unseen data.

Instead of relying on a single train-test split, the dataset is divided into multiple folds. The model is trained and validated repeatedly using different subsets, and the average performance is calculated.

This approach provides a more reliable estimate of model performance and helps reduce bias caused by a single data split.

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

rf_cv = RandomForestClassifier(random_state=42)

cv_scores = cross_val_score(
    rf_cv,
    X_train_smote,
    y_train_smote,
    cv=5,
    scoring="accuracy"
)

print("Cross Validation Scores:")
print(cv_scores)

print("\nAverage Cross Validation Accuracy:")
print(round(cv_scores.mean() * 100, 2), "%")
```

### Interpretation

The Random Forest classifier was evaluated using 5-fold Cross Validation to assess its generalization capability.

The cross-validation scores remain consistent across different folds, indicating that the model is stable and does not suffer from significant performance variation.

The average cross-validation accuracy provides a more reliable estimate of real-world model performance than a single train-test split.

# Question 7: Hyperparameter Tuning

## Task

Optimize the performance of the Random Forest classifier by selecting the best combination of hyperparameters using GridSearchCV.

---

## Explanation

Hyperparameters control the behaviour of a machine learning algorithm before the learning process begins.

GridSearchCV systematically evaluates different combinations of hyperparameters using cross validation and selects the combination that produces the best predictive performance.

Optimizing hyperparameters can improve model accuracy, robustness, and generalization ability.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5]
}

grid_search = GridSearchCV(
    RandomForestClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)

grid_search.fit(X_train_smote, y_train_smote)

print("Best Parameters:")
print(grid_search.best_params_)

print("\nBest Cross Validation Accuracy:")
print(round(grid_search.best_score_ * 100, 2), "%")
```

### Interpretation

GridSearchCV successfully identified the optimal combination of Random Forest hyperparameters.

The tuned model achieved the highest cross-validation accuracy among the evaluated parameter combinations, demonstrating that hyperparameter optimization can further improve model performance and generalization.

Compare the performance of all machine learning models using their evaluation metrics and identify the most suitable classifier for semiconductor manufacturing process classification.
